```python
def fun1(x):
    return np.power(x,2) + 5
x1 = np.linspace(-1,1)
plt.plot(x1,fun1(x1))
plt.title('X^2+5')
plt.show()
plt.title('X^2+5')

x2 = np.linspace(-6,6)
plt.plot(x2,fun1(x2))
plt.show()

x3 = np.linspace(0,5)
plt.title('X^2+5')
plt.plot(x3,fun1(x3))
plt.show()
```
