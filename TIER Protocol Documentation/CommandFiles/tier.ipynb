```python
import pandas as pd


D_PATH = "../AnalysisData/" # file path


df = pd.read_csv(f"{D_PATH}drinks.csv") #read csv



df = pd.melt(df, id_vars = ['country', 'total_litres_of_pure_alcohol'], value_vars=list(df.columns)[1:4], var_name='Drink', value_name='Quantity' ) # organizing data
df = df.sort_values(by='country') #data sorting
df.to_csv('../AnalysisData/pew_drinks.csv', index=False) #saving data 

df




```
